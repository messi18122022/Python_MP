# Übersicht der verschiedenen Plottypen in Matplotlib

Hier findest du die wichtigsten Plottypen, wie sie verwendet werden und wofür sie geeignet sind.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1. **Linienplot** (`plt.plot()`)
Ein Linienplot eignet sich für Zeitreihen oder Daten, bei denen Werte in einer kontinuierlichen Reihenfolge dargestellt werden.

### Beispiel:


```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="sin(x)", color="blue", linestyle="--")
plt.title("Linienplot")
plt.xlabel("x-Werte")
plt.ylabel("y-Werte")
plt.legend()
plt.show()
```

## 2. **Balkendiagramm** (`plt.bar()`)
Ein Balkendiagramm zeigt kategoriale Daten. Es eignet sich für den Vergleich von Werten zwischen Gruppen.

### Beispiel:


```python
kategorien = ["A", "B", "C", "D"]
werte = [10, 15, 7, 12]

plt.figure(figsize=(8, 5))
plt.bar(kategorien, werte, color="orange")
plt.title("Balkendiagramm")
plt.xlabel("Kategorien")
plt.ylabel("Werte")
plt.show()
```

## 3. **Histogramm** (`plt.hist()`)
Ein Histogramm zeigt die Verteilung numerischer Daten. Es gruppiert die Daten in Bins.

### Beispiel:


```python
daten = np.random.randn(1000)  # Zufallsdaten

plt.figure(figsize=(8, 5))
plt.hist(daten, bins=20, color="green", edgecolor="black")
plt.title("Histogramm")
plt.xlabel("Werte")
plt.ylabel("Häufigkeit")
plt.show()
```

## 4. **Streudiagramm** (`plt.scatter()`)
Ein Streudiagramm zeigt die Beziehung zwischen zwei Variablen. Es eignet sich gut, um Muster oder Korrelationen zu erkennen.

### Beispiel:


```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="purple")
plt.title("Streudiagramm")
plt.xlabel("x-Werte")
plt.ylabel("y-Werte")
plt.show()
```

## 5. **Kuchendiagramm** (`plt.pie()`)
Ein Kuchendiagramm zeigt Anteile einer Gesamtheit. Es ist nützlich für kategoriale Daten, um Prozente darzustellen.

### Beispiel:


```python
labels = ["A", "B", "C", "D"]
werte = [30, 20, 40, 10]

plt.figure(figsize=(8, 5))
plt.pie(werte, labels=labels, autopct="%1.1f%%", startangle=90, colors=["red", "blue", "green", "yellow"])
plt.title("Kuchendiagramm")
plt.show()
```

## 6. **Boxplot** (`plt.boxplot()`)
Ein Boxplot zeigt die Verteilung von Daten, einschließlich Median, Quartilen und Ausreißern.

### Beispiel:


```python
daten = [np.random.randn(100) for _ in range(4)]

plt.figure(figsize=(8, 5))
plt.boxplot(daten, labels=["A", "B", "C", "D"])
plt.title("Boxplot")
plt.ylabel("Werte")
plt.show()
```

## 7. **Heatmap** (mit `plt.imshow()`)
Eine Heatmap zeigt eine Matrix oder 2D-Daten visuell an, indem Farbschattierungen verwendet werden.

### Beispiel:


```python
matrix = np.random.rand(10, 10)

plt.figure(figsize=(8, 5))
plt.imshow(matrix, cmap="viridis", aspect="auto")
plt.colorbar(label="Werte")
plt.title("Heatmap")
plt.show()
```

## 8. **Flächendiagramm** (`plt.fill_between()`)
Ein Flächendiagramm zeigt Trends über Zeit oder Kategorien und füllt den Bereich unter der Kurve.

### Beispiel:


```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = y1 + 0.5

plt.figure(figsize=(8, 5))
plt.fill_between(x, y1, y2, color="lightblue", alpha=0.5)
plt.title("Flächendiagramm")
plt.xlabel("x-Werte")
plt.ylabel("y-Werte")
plt.show()
```

## 9. **3D-Diagramme** (mit `mpl_toolkits.mplot3d`)
Matplotlib kann auch 3D-Diagramme erstellen, z. B. Oberflächen- oder Streudiagramme.

### Beispiel:


```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(x, y, z, cmap="viridis")
plt.title("3D-Oberflächendiagramm")
plt.show()
```
